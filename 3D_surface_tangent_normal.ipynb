## **Visualización 3D de superficie, plano tangente y recta normal**

Este código en Python tiene como objetivo visualizar gráficamente en 3D una superficie definida por la función $f(x, y)=x^2+y^2$, junto con su plano tangente y la recta normal en un punto específico de la superficie.

**Se realiza lo siguiente:**  

1. **Definición de la función:**Se define la función $f(x, y)=x^2+y^2$, que representa una parábola en 3D con forma de cuenco.

2. **Generación de malla de puntos:** Se crea una malla de valores para $x$ e $y$, y se evalúa la función para obtener los valores de $z$, que definen la superficie.

3. **Cálculo del plano tangente:** Utilizando las derivadas parciales de la función, se construye la ecuación del plano tangente en el punto $(1,1,2)$. Este plano representa la mejor aproximación lineal a la superficie en ese punto.

4. **Construcción de la recta normal:** Se genera la ecuación paramétrica de la recta normal, usando como dirección el vector normal derivado del gradiente de la función en el punto de interés.

5. **Visualización en 3D:** Finalmente, se utiliza la biblioteca matplotlib para graficar la superficie, el plano tangente y la recta normal en un sistema de coordenadas tridimensional. Cada elemento se representa con diferentes colores y niveles de transparencia para facilitar su interpretación.  


Este tipo de visualización es útil en cursos de cálculo multivariable, geometría diferencial o para fines educativos y de investigación, ya que permite comprender de forma gráfica conceptos fundamentales como derivadas parciales, gradiente, plano tangente y vectores normales.

```python
# Cargando las librerías
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función f(x, y) = x^2 + y^2
def f(x, y):
    return x**2 + y**2

# Crear una malla de puntos para la superficie
gx, gy = np.meshgrid(np.linspace(-2, 2, 100), np.linspace(-2, 2, 100))
gz = f(gy, gx)

# Punto de interés donde se calcula la normal y el plano tangente
x0, y0, z0 = 1, 1, 2

# Dirección del vector normal al plano tangente
normal_dir = np.array([2, 2, -1])

# Parámetro t para graficar la recta normal
t = np.linspace(-2, 2, 20)
line_x = x0 + normal_dir[0] * t
line_y = y0 + normal_dir[1] * t
line_z = z0 + normal_dir[2] * t

# Definición del plano tangente en torno al punto (1,1)
px, py = np.meshgrid(np.linspace(0, 2, 10), np.linspace(0, 2, 10))
pz = z0 + 2 * (px - x0) + 2 * (py - y0)

# Crear la figura 3D
fig = plt.figure(figsize=(16, 12))
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
ax.plot_surface(gy, gx, gz, cmap='viridis', alpha=0.6)

# Graficar la recta normal
ax.plot(line_x, line_y, line_z, color='red', linewidth=2, label='Normal')
ax.scatter(x0, y0, z0, color='black', s=50, label='Punto (1,1,2)')

# Graficar el plano tangente
ax.plot_surface(px, py, pz, color='cyan', alpha=0.5)

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar leyenda
ax.legend()

# Mostrar el gráfico
plt.show()
```
